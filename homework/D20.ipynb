```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#1.
sns.set_style("whitegrid")
rs = np.random.RandomState(7)
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)
sns.residplot(x, y)
```
