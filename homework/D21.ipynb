```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
df = sns.load_dataset('titanic')
#2.
g = sns.FacetGrid(data= df,col='sex',row='survived' )
g.map(plt.scatter, 'sex', 'survived' )
```
